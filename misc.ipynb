```python
# Import necessary packages

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Matplotlib Style Settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# Comparing different activation functions

def ReLU(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.linspace(-5, 5, 1000)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.axhline(y=0, color='k', lw=1)
ax.axvline(x=0, color='k', lw=1)
ax.plot(x, ReLU(x), color='navy')

ax.margins(x=0, y=0.1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
fig.savefig('./figures/ReLU.png')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.axhline(y=0, color='k', lw=1)
ax.axhline(y=1, color='k', lw=1, ls=':')
ax.axvline(x=0, color='k', lw=1)
ax.plot(x, sigmoid(x), color='navy')

ax.margins(x=0, y=0.1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([0, 1])
plt.show()
```

```python
fig.savefig('./figures/sigmoid.png')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.axhline(y=0, color='k', lw=1)
ax.axhline(y=1, color='k', lw=1, ls=':')
ax.axhline(y=-1, color='k', lw=1, ls=':')
ax.axvline(x=0, color='k', lw=1)
ax.plot(x, tanh(x), color='navy')

ax.margins(x=0, y=0.1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_xticks([])
ax.set_yticks([-1, 0, 1])
plt.show()
```

```python
fig.savefig('./figures/tanh.png')
```
