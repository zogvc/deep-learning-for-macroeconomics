```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.unicode_minus'] = False
import seaborn as sns

from fredapi import Fred
```

```python
KEY = '422cad069f7f276da5d9b8f9b00480b3'
fred = Fred(api_key=KEY)
```

```python
start = '1969-11-01' 
end = '2024-04-30'
```

```python
ipi = fred.get_series('INDPRO', observation_start=start, observation_end=end)   # Industrial Production Index : proxy for economic output
pce = fred.get_series('PCE', observation_start=start, observation_end=end)      # Personal Consumption Index  : proxy for inflation
ffr = fred.get_series('FEDFUNDS', observation_start=start, observation_end=end) # Federal Funds Rate          : proxy for interest rate
uer = fred.get_series('UNRATE', observation_start=start, observation_end=end)   # Unemployment Rate           : proxy for labour market indicators
```

```python
# Visualization of Raw Data

fig, axs = plt.subplots(2, 2, figsize=(10, 7))
axs = axs.flatten()

axs[0].plot(ipi, color='navy')
axs[0].set_title('Industrial Production Index', fontsize=11)

axs[1].plot(pce, color='navy')
axs[1].set_title('Personal Consumption Expenditure', fontsize=11)

axs[2].plot(ffr, color='navy')
axs[2].set_title('Federal Funds Effective Rate', fontsize=11)

axs[3].plot(uer, color='navy')
axs[3].set_title('Unemployment Rate', fontsize=11)

for ax in axs:
    ax.tick_params(direction='in')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid( color='lightgray')
    #ax.set_xticks(pd.date_range('1970', '2030', freq='D'), fontsize=10)

#plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.3)
plt.show()
```

```python
fig.savefig('./figures/raw_viz.png')
```

```python
pd.concat([ipi.rename('ipi'), pce.rename('pce'), ffr.rename('ffr'), uer.rename('uer')], axis=1).describe()
```

```python
# Preprocessing

ipi1 = np.log(ipi).diff()
pce1 = np.log(pce) - 2 * np.log(pce).shift(1) + np.log(pce).shift(2)
ffr1 = ffr.diff()
uer1 = uer.diff()
```

```python
# Visualization of Processed Data

fig, axs = plt.subplots(2, 2, figsize=(10, 7))
axs = axs.flatten()

axs[0].plot(ipi1[2:], color='navy')
axs[0].set_title('Industrial Production Index', fontsize=11)

axs[1].plot(pce1[2:], color='navy')
axs[1].set_title('Personal Consumption Expenditure', fontsize=11)

axs[2].plot(ffr1[2:], color='navy')
axs[2].set_title('Federal Funds Effective Rate', fontsize=11)

axs[3].plot(uer1[2:], color='navy')
axs[3].set_title('Unemployment Rate', fontsize=11)

for ax in axs:
    ax.tick_params(direction='in')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid( color='lightgray')
    #ax.set_xticks(pd.date_range('1970', '2030', freq='D'), fontsize=10)

#plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.3)
plt.show()
```

```python
fig.savefig('./figures/processed_viz.png')
```

```python
# Concatenate all 4 variables.
df = pd.concat([ipi1[2:].rename('ipi'), pce1[2:].rename('pce'), ffr1[2:].rename('ffr'), uer1[2:].rename('uer')], axis=1)
df.index = pd.date_range('1970-01-01', '2024-05-01', freq='M')
df = df[['ipi', 'uer', 'pce', 'ffr']]
```

```python
df.to_csv('./data/macro_processed.csv')
```
