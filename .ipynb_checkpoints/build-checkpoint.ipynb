```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Check the data length.
# Optional

data_length = []

for yr in range(2016, 2024):
    df = pd.read_csv(f'./data/venture{yr}.csv', encoding='cp949')
    data_length.append(len(df))

print(data_length)
```

```python
# Import data by year.
df16 = pd.read_csv('./data/venture2016.csv', encoding='cp949')
df17 = pd.read_csv('./data/venture2017.csv', encoding='cp949')
df18 = pd.read_csv('./data/venture2018.csv', encoding='cp949')
df19 = pd.read_csv('./data/venture2019.csv', encoding='cp949')
df20 = pd.read_csv('./data/venture2020.csv', encoding='cp949')
df21 = pd.read_csv('./data/venture2021.csv', encoding='cp949')
df22 = pd.read_csv('./data/venture2022.csv', encoding='cp949')
df23 = pd.read_csv('./data/venture2023.csv', encoding='cp949')

# Drop duplicated columns.
df21.drop(df21.columns[df21.columns.str[-2:] == '.1'], axis=1, inplace=True)
df22.drop(df22.columns[df22.columns.str[-2:] == '.1'], axis=1, inplace=True)
df23.drop(df23.columns[df23.columns.str[-2:] == '.1'], axis=1, inplace=True)

data = [df16, df17, df18, df19, df20, df21, df22, df23]
```

```python
df23.head(10)
```

```python
# Construct a dataframe of columns and save it as excel file.
# Optional. 

cols = pd.DataFrame()

for i, df in enumerate(data):
    cols = pd.concat([cols, pd.Series(df.columns.rename(2016+i))], axis=1)
    
cols.to_excel('./data/columns.xlsx')
```
